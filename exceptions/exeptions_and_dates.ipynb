## Задание 1
Печатные газеты использовали свой формат дат для каждого выпуска. Для каждой газеты из списка напишите формат указанной даты для перевода в объект datetime:

- `The Moscow Times - Wednesday, October 2, 2002`
- `The Guardian - Friday, 11.10.13`
- `Daily News - Thursday, 18 August 1977`

```python
the_moscow_time_format = '%A, %B %d, %Y'
the_guardian_format = '%A, %d.%m.%y'
daily_news_format = '%A, %d %B %Y'

task = [
'The Moscow Times - Wednesday, October 2, 2002',
'The Guardian - Friday, 11.10.13',
'Daily News - Thursday, 18 August 1977']

answer = [
    the_moscow_time_format,
    the_guardian_format,
    daily_news_format
]

for i, line in enumerate(task):
    name, date_str = line.split(' - ')
    task_date_ = datetime.strptime(date_str, answer[i]).date()

    print(f'{name} - {task_date_.strftime("%d %B %Y")}')
```

## Задание 2

Дан поток дат в формате `YYYY-MM-DD`, в которых встречаются некорректные значения:

`stream = ['2018-04-02', '2018-02-29', '2018-19-02']`

Напишите функцию, которая проверяет эти даты на корректность. Т. е. для каждой даты возвращает True (дата корректна) или False (некорректная дата).

```python
def check_str_date(str_date):
    try:
        y, m, d = map(int, str_date.split('-'))
        date(y, m, d)
    except ValueError:
        return False
    return True

stream = ['2018-04-02', '2018-02-29', '2018-19-02']
for s in stream:
    print(f'{s} -- {"Correct" if check_str_date(s) else "Wrong"}')
    
```

## Задание 3

Напишите функцию `date_range`, которая возвращает список дат за период от `start_date` до `end_date`. Даты должны вводиться в формате `YYYY-MM-DD`. В случае неверного формата или при `start_date > end_date` должен возвращаться пустой список.

```python
def date_list(start_date: str, end_date: str) -> list:
    try:
        s_date, e_date = map(
            lambda s: datetime.strptime(s, '%Y-%m-%d').date(),
            (start_date, end_date))
    except (ValueError, TypeError):
        return []
    result = []
    t_day = timedelta(days=1)
    while s_date <= e_date:
        result.append(s_date)
        s_date += t_day
    return result

assert(len(date_list('2020-02-01', '2020-03-01')) == 30)
assert(len(date_list('2020-2-01', '2020-3-01')) == 30)
assert(len(date_list('2019-02-01', '2019-03-01')) == 29)
assert(len(date_list('2011-01-01', '2011-01-01')) == 1)
assert(len(date_list('2011-01-02', '2011-01-01')) == 0)
assert(len(date_list('2020-021-01', '2020-03-01')) == 0)
assert(len(date_list('2020-02-01', '2020-13-01')) == 0)
assert(len(date_list('2020-02-30', '2020-12-01')) == 0)
```
