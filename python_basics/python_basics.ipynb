## Задание 1

Даны 2 переменных, в которых хранятся строки произвольной длины: *phrase_1* и *phrase_2*.  
Напишите код, который проверяет какая из этих строк длиннее.

Примеры работы программы:  
1.  

```
phrase_1 = 'Насколько проще было бы писать программы, если бы не заказчики'
phrase_2 = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'

```

Результат:  
`Фраза 1 длиннее фразы 2`
2.     

```
phrase_1 = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'
phrase_2 = 'Насколько проще было бы писать программы, если бы не заказчики'
```

Результат:  
`Фраза 2 длиннее фразы 1`
3.     

```
phrase_1 = 'Насколько проще было бы писать программы, если бы не заказчики'
phrase_2 = 'Насколько проще было бы писать программы, если бы не заказчики'
```

Результат:  
`Фразы равной длины`

````python
def comparePhrases(ph1: str, ph2: str) -> int:
    len1 = len(ph1)
    len2 = len(ph2)
    if len1 > len2:
        return 1
    elif len1 < len2:
        return 2
    return 0

def task1(phrase_1: str, phrase_2: str) -> str:
    res = comparePhrases(phrase_1, phrase_2)
    return [
        'Фразы равной длины',
        'Фраза 1 длиннее фразы 2',
        'Фраза 2 длиннее фразы 1'
    ][res]

assert(task1(
    'Насколько проще было бы писать программы, если бы не заказчики',
    '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)') 
       == 'Фраза 1 длиннее фразы 2')
assert(task1(
    '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)', 
    'Насколько проще было бы писать программы, если бы не заказчики')
       == 'Фраза 2 длиннее фразы 1')
assert(task1(
    'Насколько проще было бы писать программы, если бы не заказчики',
    'Насколько проще было бы писать программы, если бы не заказчики') 
       == 'Фразы равной длины')
print('Решение Задания 1')
````

## Задание 2

Дана переменная, в которой хранится четырехзначное число (год). Необходимо написать программу, которая выведет, является ли данный год високосным или обычным.

Пример работы программы:
1.  

```
year = 2020
```
Результат:  
`Високосный год` 

2.  

```
year = 2019
```
Результат:  
`Обычный год` 

````python
# год, номер которого кратен 400, — високосный;
# остальные годы, номер которых кратен 100, — невисокосные (например, годы 1700, 1800, 1900, 2100, 2200, 2300);
# остальные годы, номер которых кратен 4, — високосные
def isLeapYear(year: int) -> bool:
    return bool(not year % 400 or (not year % 4 and year % 100))

def task2(year: int) -> str:
    return [
        'Обычный год',
        'Високосный год'
    ][isLeapYear(year)]

assert(task2(2020) == 'Високосный год')
assert(task2(2019) == 'Обычный год')
assert(task2(2000) == 'Високосный год')
assert(task2(1900) == 'Обычный год')
assert(task2(1804) == 'Високосный год')
assert(task2(1800) == 'Обычный год')
print('Решение Задания 2')
````

## Задание 3

Необходимо написать программу, которая будет запрашивать у пользователя месяц и дату рождения и выводить соответствующий знак зодиака.

Пример работы программы:  
1.  

```
Введите день:
30

Введите месяц:
Август
```
Результат:  
`Ваш знак зодиака: Дева` 

2.  

```
Введите день:
29

Введите месяц:
Октябрь
```
Результат:  
`Ваш знак зодиака: Скорпион` 


````python
# Не проверяем насколько верно введен день в месяце
# Данные по датам взяты отсюда
# https://en.wikipedia.org/wiki/Astrological_sign#Dates_table
SIGNS_PER_MONTH = [
    [19, 'Козерог', 'Водолей'], # Январь
    [18, 'Водолей', 'Рыбы'], # Февраль
    [20, 'Рыбы', 'Овен'], # Март
    [19, 'Овен', 'Телец'], # Апрель
    [20, 'Телец', 'Близнецы'], # Май
    [22, 'Близнецы', 'Рак'], # Июнь
    [22, 'Рак', 'Лев'], # Июль
    [22, 'Лев', 'Дева'], # Август
    [22, 'Дева', 'Весы'], # Сентябрь
    [22, 'Весы', 'Скорпион'], # Октябрь
    [21, 'Скорпион', 'Стрелец'], # Ноябрь
    [21, 'Стрелец', 'Козерог'] # Декабрь
]
MONTH_NAMES = [
    'январь', 'февраль','март', 'апрель','май','июнь',
    'июль', 'август', 'сентябрь', 'октябрь', 'ноябрь', 'декабрь'
]

def detectSign(day: int, month: str) -> str:
    month = month.lower()
    if month not in MONTH_NAMES:
        return None
    month_num = MONTH_NAMES.index(month)
    try:
        day = int(day)
    except ValueError:
        return None
    if day <= 0 or day > 31:
        return None
    last_day, prev_sign, next_sign = SIGNS_PER_MONTH[month_num]
    if day > last_day:
        return next_sign
    else:
        return prev_sign

def task3():
    day = input('Введите день:\n')
    month = input('Введите месяц:\n')
    sign = detectSign(day, month)
    if sign is None:
        print('Уверены что верно ввели данные?')
    else:
        print('Ваш знак зодиака: {}'.format(sign))

task3()
````

## Задание 4

Вам нужно написать программу для подбора упаковок по размерам товара. Размеры хранятся в переменных (в сантиметрах):

```
width = 10
length = 205
height = 5
```

Используйте следующие правила:
- если каждое из трех измерений менее 15 сантиметров, то выведите на экран "Коробка №1";
- если хотя бы одно из измерений больше 15 сантиметров, но менее 50 сантиметров, то выводите "Коробка №2";
- если длина товара больше 2 метров, то выводите "Упаковка для лыж";
- во всех остальных случаях выводите "Стандартная коробка №3".


````python
# Тут много вопросов в каком порядке проверять, особенно про вторую коробку
def chooseBox(width: int, length: int, height: int) -> int:
    dims = [width, length, height]
    if all(d <= 15 for d in dims):
        return 0
    if any(15 < d <= 50 for d in dims):
        return 1
    if length > 200:
        return 2
    return 3

def task4(width: int, length: int, height: int) -> str:
    return [
        "Коробка №1",
        "Коробка №2",
        "Упаковка для лыж",
        "Стандартная коробка №3"][chooseBox(width, length, height)]

assert(task4(10, 10, 14) == "Коробка №1")
assert(task4(40, 10, 14) == "Коробка №2")
assert(task4(51, 250, 55) == "Упаковка для лыж")
assert(task4(140, 110, 114) == "Коробка №3")

print('Решение Задания 4')
````
