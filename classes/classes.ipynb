## Задание 1

Напишите функцию, которая возвращает название валюты (поле ‘Name’) с максимальным значением курса с помощью сервиса https://www.cbr-xml-daily.ru/daily_json.js

```python
import exchange

def get_max_rate():
    rates = exchange.Rate().exchange_rates()
    max_rate = 0
    max_rate_name = ''
    for code, info in rates.items():
        rate = info.get('Value', 0)
        if rate > max_rate:
            max_rate = info['Value']
            max_rate_name = info.get('Name', '')
    return max_rate_name

get_max_rate()
```

## Задание 2

Добавьте в класс `Rate` параметр `diff` (со значениями True или False), который в случае значения True в методах курсов валют (eur, usd итд) будет возвращать не курс валюты, а изменение по сравнению в прошлым значением. Считайте, self.diff будет принимать значение True только при возврате значения курса. При отображении всей информации о валюте он не используется.

```python
class Rate(exchange.Rate):
    def __init__(self, diff, **kwargs):
        super().__init__(**kwargs)
        self.diff = diff
    
    def _get_diff(self, currency):
        response = self.exchange_rates()
        data = response.get(currency)
        if not data:
            return f'Error: Unknown currency {currency}'
        prev_rate = data.get('Previous')
        if prev_rate is None: # Safer check if Previous can be 0
            return f'Error: No previous rate for {currency}'
        current_rate = data.get('Value') 
        if current_rate is None:
            return f'Error: No current rate for {currency}'
        return current_rate - prev_rate
    
    def eur(self):
        if self.diff:
            return self._get_diff('EUR')
        return self.make_format('EUR')
    
    def usd(self):
        if self.diff:
            return self._get_diff('USD')
        return self.make_format('USD')

    def AZN(self):
        if self.diff:
            return self._get_diff('AZN')
        return self.make_format('AZN')

rate = Rate(True)
print(f'USD diff {rate.usd()}')
print(f'EUR diff {rate.eur()}')
```

```python
class Employee:
    def __init__(self, name, seniority):
        self.name = name
        self.seniority = seniority
        self.grade = 1
    
    def grade_up(self):
        """Повышает уровень сотрудника"""
        self.grade += 1
    
    def publish_grade(self):
        """Публикация результатов аккредитации сотрудников"""
        print(self.name, self.grade)
    
    def check_if_it_is_time_for_upgrade(self):
        pass

class Designer(Employee):
    def __init__(self, name, seniority):
        super().__init__(name, seniority)
        self.awards = 2
    
    def check_if_it_is_time_for_upgrade(self):
        # для каждой аккредитации увеличиваем счетчик на 1
        self.seniority += 1
        
        points = self.seniority + self.awards * 2
        grade_by_points = points // 7 + 1
        if grade_by_points < self.grade:
            print('Oops, have you changed the grade directly? Call the admin.')
        else:
            for _ in range(self.grade, grade_by_points):
                self.grade_up()
        # публикация результатов
        return self.publish_grade()
```
