Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

Исходные данные имеют следующую структуру:

1. перечень всех документов
```
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```
2. перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
```
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

Общие требования к программе:
- код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
- в коде отсутствуют глобальные переменные (за исключением **documents** и **directories**);
- пользовательский ввод обрабатывается в цикле **while** до тех пор, пока пользователь явно не завершит программу (вводом команды "**q**").

````python
# Defining global data
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
````

### Пункт 5. Пользователь по команде "*ds*" может удалить существующую полку из данных (только если она пустая)

Примеры работы:

1.

```
Введите команду:
ds

Введите номер полки:
3
```
Результат:  
`Полка удалена. Текущий перечень полок: 1, 2.`

2.
```
Введите команду:
ds

Введите номер полки:
1
```
Результат:  
`На полке есть документа, удалите их перед удалением полки. Текущий перечень полок: 1, 2, 3.`

3.
```
Введите команду:
ds

Введите номер полки:
4
```
Результат:  
`Такой полки не существует. Текущий перечень полок: 1, 2, 3.`

````python
class InputDone(Exception): pass

def get_doc(doc_id):
    global documents # HW: It's not necessary, just to expain where it's from
    for doc in documents:
        if 'number' in doc and doc['number'] == doc_id:
            return doc
    return None

def get_full_doc_info(doc):
    # TODO: Well, it's better to check input doc of course
    directory = get_doc_directory(doc['number'])
    result = doc.copy()
    result['directory'] = directory
    return result

def get_doc_owner(doc_id):
    doc = get_doc(doc_id)
    return doc and (doc['name'] if 'name' in doc else 'Неизвестный')

def get_doc_directory(doc_id):
    # To be sure we check if such doc exists (even if dirs have the id)
    if not get_doc(doc_id):
        return None
    global directories
    for dir_n, docs in directories.items():
        if doc_id in docs:
            return dir_n
    return None


def quit():
    raise InputDone

def print_doc_owner():
    owner = get_doc_owner(input('Введите номер документа:\n'))
    if owner:
        print(f'Владелец документа: {owner}')
    else:
        print('Документ не найден в базе')

def print_doc_directory():
    directory = get_doc_directory(input('Введите номер документа:\n'))
    if directory:
        print(f'Документ хранится на полке:: {directory}')
    else:
        print('Документ не найден в базе')

def print_all_docs():
    global documents
    for doc in documents:
        di = get_full_doc_info(doc)
        print(f'№: {di["number"]}, тип: {di["type"]}, владелец: {di["name"]}, полка хранения: {di["directory"]}')

def remove_directory():
    dir_id = input('Введите номер полки:\n')
    global directories
    dir_names = ', '.join(directories.keys())
    if dir_id in directories:
        if len(directories[dir_id]):
            print(f'На полке есть документа, удалите их перед удалением полки. Текущий перечень полок: {dir_names}.')
        else:
            directories.pop(dir_id)
            dir_names = ', '.join(directories.keys())
            print(f'Полка удалена. Текущий перечень полок: {dir_names}.')
    else:
        print(f'Такой полки не существует. Текущий перечень полок: {dir_names}.')

def add_directory():
    dir_id = input('Введите номер полки:\n')
    global directories
    dir_names = ', '.join(directories.keys())
    if dir_id in directories:        
        print(f'Такая полка уже существует. Текущий перечень полок: {dir_names}.')
    else:
        directories[dir_id] = []
        print(f'Полка добавлена. Текущий перечень полок: {dir_names}, {dir_id}.')
    
commands = {
    'q': quit,
    'p': print_doc_owner,
    's': print_doc_directory,
    'l': print_all_docs,
    'ads': add_directory,
    'ds': remove_directory
}    


while True:
    com = input('Введите команду:\n')
    if com in commands:
        try:
            commands[com]()
        except InputDone:
            print('Всего доброго!')
            break
    else:
        command_list = ', '.join(list(commands.keys()))
        print(f'Данная команда неизвестна. Попробуйте одну из: {command_list}')
````
