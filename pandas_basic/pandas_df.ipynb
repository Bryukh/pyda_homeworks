```python
movies = pd.read_csv('./data/ml-latest-small/movies.csv')
ratings = pd.read_csv('./data/ml-latest-small/ratings.csv')
print('MOVIES', len(movies))
print(movies.head())
print('RATINGS', len(ratings))
print(ratings.head())
```

## Задание 2

По данным файла `power.csv` посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения `quantity`.

```python
!ls ./data/
```

```python
power = pd.read_csv('data/power.csv')
power.info()
```

```python
baltic = ['Latvia', 'Estonia', 'Lithuania']
categories = [4, 12, 21]

total_quantity = power[ (power['country'].isin(baltic))
      & (power['category'].isin(categories)) 
      & (power['year'].between(2005, 2010, True))
      & (power['quantity'] > 0)]['quantity'].sum()
print(total_quantity)
```

Ответ: Суммарное потребление стран Прибалтики категорий 4, 12 и 21 за период с 2005 по 2010 года (включительно) -- **240580**
